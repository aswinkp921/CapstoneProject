<h3><b>Outlier removal</b></h3>

```python
# Function to cap outliers at the 5th and 95th percentiles
def cap_outliers(df, columns):
    for column in columns:
        lower_bound = df[column].quantile(0.05)
        upper_bound = df[column].quantile(0.95)
        df[column] = df[column].clip(lower=lower_bound, upper=upper_bound)
    return df

# Columns to handle outliers
columns_to_cap = ['PTS', 'AST', 'TRB', 'MP', 'PER', 'TS%', '3PAr', 'FTr', 'WS/48']

# Cap the outliers in the selected columns
data_capped = cap_outliers(data.copy(), columns_to_cap)

# Verify changes by plotting the boxplot again
plt.figure(figsize=(15, 10))
data_capped[columns_to_cap].boxplot()
plt.title('Boxplot of Selected After Capping Outliers')
plt.xticks(rotation=45)
plt.show()
```

The outlier removal technique I've implemented is known as "capping" or "trimming" outliers. Specifically, this method involves capping or limiting the outliers to a certain percentile range, the 5th and 95th percentiles. This approach is sometimes referred to as "Winsorization" (though technically Winsorization replaces the outliers with the nearest value within the percentile range, while capping just limits them).

This technique ensures that extreme values are not removed entirely from the dataset but are instead adjusted to fall within a more reasonable range, thus reducing their impact on the overall analysis while preserving the data points.

```python
from scipy.stats import yeojohnson

# Log Transformation
log_transform_features = ['ORB%', 'BLK%', '3P', '3PA', 'FT']
data_transformed = data_capped.copy()
for feature in log_transform_features:
    data_transformed[feature] = np.log1p(data_transformed[feature])

# Square Root Transformation
sqrt_transform_features = ['GS', 'DRB%', 'TRB%', 'AST%', 'TOV%', 'OWS', 'DWS', 'WS', 'VORP', 'FG']
for feature in sqrt_transform_features:
    data_transformed[feature] = np.sqrt(data_transformed[feature])

# Yeo-Johnson Transformation (handles negative and non-positive values)
yeojohnson_transform_features = ['OBPM', 'BPM']
for feature in yeojohnson_transform_features:
    data_transformed[feature], _ = yeojohnson(data_transformed[feature])

# Checking the results
transformed_skewness = data_transformed[log_transform_features + sqrt_transform_features + yeojohnson_transform_features].apply(skew)
transformed_kurtosis = data_transformed[log_transform_features + sqrt_transform_features + yeojohnson_transform_features].apply(lambda x: kurtosis(x, fisher=False))

print("Skewness after transformation:\n", transformed_skewness)
print("Kurtosis after transformation:\n", transformed_kurtosis)
```

### 

Skewness After Transformation:

- Most features now have skewness values closer to 0, indicating that the data is more symmetrically distributed after the transformations.
- Features like **ORB%**, **3P**, **3PA**, and **AST%** have skewness values near 0, which suggests a much-improved symmetry.
- Some features, such as **TOV%** and **FT**, still show some skewness, but it's significantly reduced compared to the original values.

### Kurtosis After Transformation:

- Many features now have kurtosis values closer to 3, indicating that their distribution is more normal-like.
- **Leptokurtic Features**: **TOV%**, **OBPM**, and **BPM** still exhibit high kurtosis, indicating that they have heavy tails or more extreme outliers.
- Some features like **GS** and **FG** now have kurtosis values below 3, suggesting a reduction in outliers and more normally distributed data.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, roc_curve, roc_auc_score

# Define the models
models = {
    "Linear Regression": LinearRegression(),
    "Support Vector Regression": SVR(),
    "Random Forest Regressor": RandomForestRegressor(random_state=42),
    "Gradient Boosting Regressor": GradientBoostingRegressor(random_state=42),
    "AdaBoost Regressor": AdaBoostRegressor(random_state=42)
}

# Lists to store the evaluation metrics for each model
model_names = []
maes = []
mses = []
rmses = []
r2s = []
roc_aucs = []  # For models that support predict_proba

# Train and evaluate each model
for model_name, model in models.items():
    model.fit(X_train_reduced, y_train)  # Use the reduced data
    y_pred = model.predict(X_test_reduced)
    
    model_names.append(model_name)
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = mean_squared_error(y_test, y_pred, squared=False)
    r2 = r2_score(y_test, y_pred)
    
    maes.append(mae)
    mses.append(mse)
    rmses.append(rmse)
    r2s.append(r2)
    
    print(f"Model: {model_name}")
    print("Mean Absolute Error (MAE):", mae)
    print("Mean Squared Error (MSE):", mse)
    print("Root Mean Squared Error (RMSE):", rmse)
    print("R² Score:", r2)
    
    if hasattr(model, "predict_proba") or hasattr(model, "decision_function"):
        if hasattr(model, "predict_proba"):
            y_pred_prob = model.predict_proba(X_test_reduced)[:, 1]
        else:
            y_pred_prob = model.decision_function(X_test_reduced)
        
        fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
        roc_auc = roc_auc_score(y_test, y_pred_prob)
        roc_aucs.append(roc_auc)
        
        # Plot ROC curve
        plt.figure()
        plt.plot(fpr, tpr, color='red', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
        plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(f'ROC Curve for {model_name}', color='orange')
        plt.legend(loc="lower right")
        plt.show()
        print(f"ROC curve for {model_name} plotted successfully.")
    else:
        roc_aucs.append(None)
        print(f"{model_name} does not support ROC curve plotting.")
    
    print("\n")

# Summarize the evaluation metrics for all models in a DataFrame
results_df = pd.DataFrame({
    'Model-Name': model_names,
    'MAE': maes,
    'MSE': mses,
    'RMSE': rmses,
    'R² Score': r2s,
    'ROC AUC': roc_aucs
})

# Display the DataFrame to check the results
print(results_df)
```

<h2><b>Limitations</b></h2>


### Lack of Comprehensive Performance Metrics:
While the models accurately predict player performance based on available metrics, the dataset may not include all relevant performance indicators, such as advanced analytics or off-court factors that could impact a player's game.

### Potential Overfitting:
The Random Forest model performed exceptionally well, but there is a risk of overfitting to the training data, especially given the complexity of the model and the number of features used. This may limit the model's ability to generalize to new, unseen data.

### Feature Redundancy:
Although feature selection techniques were applied, there may still be redundant features within the dataset, which could lead to multicollinearity issues and affect the model's interpretability and performance.

### Inadequate Representation of Rare Events:
The dataset may not sufficiently represent rare but impactful events, such as exceptionally high or low player performances. This could limit the model's ability to predict these outlier scenarios accurately.

### Impact of Data Imputation:
Handling missing values through imputation might introduce bias, especially if the missing data patterns are not random. This could affect the model's predictions and reliability.

### Player Variability:
The model does not account for the variability in player performance across different teams, seasons, or match conditions. External factors like injuries, team dynamics, or coaching strategies are not included in the dataset, which could lead to less accurate predictions.

### Complexity of the NBA Environment:
The NBA is a dynamic environment with many variables that influence player performance. The models may oversimplify these complexities, potentially leading to predictions that do not fully capture the real-world scenarios.

### Dependence on Historical Data:
The models rely heavily on historical data, which may not fully account for changes in a player's style, team strategies, or league-wide trends over time. This could affect the model's ability to predict future performance accurately.
